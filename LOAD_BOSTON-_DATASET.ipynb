```python
#now apply linear regression
def hypothesis(X,theta):
    y_ = 0
    m = X.shape[0]
    for i in range(m):
        y_ = theta[i]*X[i]
        
    return y_    
```

```python
def cost(X,Y, theta):
    m = X.shape[0]
    total_loss = 0
    for i in range(m):
        y_p = hypothesis(X,theta)
        
        y_gt = Y[i]
        
        total_loss += (y_p - y_gt)**2
        
    return total_loss/m    
```

```python
def gradient(X ,Y, theta):
    m ,n =X.shape
    grad = np.zeros((n,))
    
    for j in range(n):
        for i in range(m):
            y_p = hypothesis(X,theta)
            
            y_gt = Y[i]
            
            grad[j] += (y_p - y_gt)*X[i][j]
            
    return grad/m        
    
    
```

```python
def gradient_descent(X,Y, learning_rate = 0.001, epochs = 1000):
    m, n = X.shape
    theta = np.zeros((n,))
    cost_epoch = []
    
    for i in range(epochs):
        
        loss = cost(X,Y,theta)
        grad = gradient(X,Y,theta)
        cost_epoch.append(loss)
        
        for j in range(n):
            
            theta[j] = theta[j] -learning_rate*grad[j]
            
    return theta ,cost_epoch       
```

```python
theta ,cost_epoch = gradient_descent(X,Y,learning_rate = 0.001, epochs =1000)
```
