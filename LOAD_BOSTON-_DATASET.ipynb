```python
def hypothesis(x, theta):
    
    y_ = 0
    n = x.shape[0]
    
    for i in range(n):
        y_ += theta[i]*x[i]
        
    return y_  
```

```python
def cost(x, y ,theta):
    
    m = x.shape[0]
    
    total_loss = 0
    for i in range(m):
        
        y_p = hypothesis(x, theta)
        
        y_gt = y[i]
        
        total_loss += (y_p - y_gt)**2
        
    return total_loss/m   
```

```python
def gradient(x, y, theta):
    m, n = x.shape
    
    grad = np.zeros((2,))
    
    for j in range(n):
        for i in range(m):
            
            y_p = hypothesis(x[i], theta)
            
            yx_gt = y[i]
            
            grad[j] += (y_p - y_gt)*x[i][j]
            
    return grad/m       
```

```python
def gradient_descent(x, y, learning_rate = 0.001, epochs = 100):
    
    m, n =x.shape
    theta = np.zeros((n,))
    cost_epoch = []
    
    for i in range(epochs):
        loss = cost(x, y, theta)
        grad = gradient(x, y, theta)
        cost_epoch.append(loss)
        
        for j in range(n):
            
            theta[j] = theta[j] -learning_rate*grad[j]
            
    return theta, cost_epoch        
```
